```python
# loading required packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading dataset
df = pd.read_csv("/content/india_covid_19_state_wise_info.csv")
```

```python
# viewing the data
df.head()
```

```python
# Finding wether there are any missing values
df.isna().sum()
```

```python
del df["sl_no"]
```

# Performing EDA

**States with their respective covid cases**

```python
# Set the width and height of the figure
plt.figure(figsize=(10,10))

# Add title
plt.title("Confirmed_Cases vs States")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(y = df["state_ut"], x = df["confirmed_cases"],order=df.sort_values('confirmed_cases',ascending=False).state_ut)

#  label for horizontal axis
plt.xlabel("States")
#  label for vertical axis
plt.ylabel("No.of cases")
```

**Which are the states most Effected by covid-19 in India**

```python
df1 = df.sort_values(by=['confirmed_cases'],ascending=False)
df1 = df1.head()
sns.barplot(y = df1["state_ut"], x = df1["confirmed_cases"])
#  label for horizontal axis
plt.ylabel("States")
#  label for vertical axis
plt.xlabel("cases")
```

**From the graph we can see that the top 5 states with active cases are:**

1.   Maharashtra
2.   Andhra Pradesh
1.   Tamil Nadu
2.   Karnataka
1.   Uttar pradesh








**Which states are with most recovered cases**

```python
df2 = df.sort_values(by=['recovered_cases'],ascending=False)
df2 = df2.head()
sns.barplot(y = df2["state_ut"], x = df2["recovered_cases"])
#  label for horizontal axis
plt.ylabel("States")
#  label for vertical axis
plt.xlabel("Recovered Cases")
```

**From the graph we can see that the top 5 states with recovered cases are:**

1.   Maharashtra
2.   Andhra Pradesh
1.   Tamil Nadu
2.   Karnataka
1.   Uttar pradesh

**which states are with most death cases**

```python
df3 = df.sort_values(by=['death_cases'],ascending=False)
df3 = df3.head()
sns.barplot(y = df3["state_ut"], x = df3["death_cases"])
#  label for horizontal axis
plt.ylabel("States")
#  label for vertical axis
plt.xlabel("Deaths")
```

**From the graph we can see that the top 5 states with recovered cases are:**

1.   Maharashtra
2.   Tamil Nadu
1.   Karnataka
2.   Andhra Pradesh
1.   Uttar pradesh

**Between confirmed cases and recovered cases**

```python
# plt.figure(figsize=(10,10))
sns.regplot(x=df["confirmed_cases"], y=df["recovered_cases"])
plt.xlabel("Confirmed Cases")
plt.ylabel("Recovered Cases")
```
